```python
from datetime import date
from bs4 import BeautifulSoup as soup
import requests,csv
```

```python
today = date.today()
d = today.strftime("%m-%d-%y")
cnn_url="https://www.daiict.ac.in/staff".format(d)
html = requests.get(cnn_url)
bsobj = soup(html.content,'lxml')
```

In above section we have create html request to the url and implemented a object variable "bsobj" which acts an instance to access the content of web page.

```python
staff_names = []
for row in bsobj.findAll("h3"):
    staff_names.append(row.text.strip())
```

In above we have created a list staff_name which stored the names of staff member though the object created bsobj in the header section "h3" of staff webpage. Here .strip() plays a role of space removal between text

```python
for person in staff_names:
  print(person)
```

Printing the names of staff members present in Daiict.

```python
staff_educations = []
for row in bsobj.findAll('div',{'class':'facultyQualification'}):
    staff_educations.append(row.text.strip())
```

In above we have created a list staff_educations which stored the education of staff member though the object created bsobj in the div section of class ="facultyQualification" of staff webpage.Here .strip() plays a role of space removal between text.

```python
for qualification in staff_educations:
  print(qualification)
```

```python
staff_roles = []
for row in bsobj.findAll('div',{'class':'facultyEducation'}):
    staff_roles.append(row.text.strip())
```

In above we have created a list staff_roles which stored the qualification of staff member though the object created bsobj in the div section of class ="facultyEducation" of staff webpage.Here .strip() plays a role of space removal between text.

```python
for Individual_role in staff_roles:
  print(Individual_role)
```

The above displays us staff qualification.

```python
staff_emails = []
for row in bsobj.findAll('span', {'class': 'facultyemail'}):
    first_email = row.text.strip().split(',')[0]
    staff_emails.append(first_email)
```

In above we have created a list staff_emails which stored the emails of staff member though the object created bsobj in the Span of class ="facultyemail" of staff webpage.Here .strip() plays a role of space removal between text. More over we have there were multiple emails presents corresponding to several staff but for simplicity we have taken only 1st instance of email of each row.

```python
for Indvidual_email in staff_emails:
  print(Indvidual_email)
```

```python
staff_Phone_number = []
for row in bsobj.findAll('span', {'class': 'facultyNumber'}):
    phone_number = str(row)[28:46]
    staff_Phone_number.append(phone_number)
```

In above we have created a list staff_Phone_number which stored the Phone_number of staff member though the object created bsobj in the Span of class ="facultynumber" of staff webpage

```python
for Indvidual_Phone_num in staff_Phone_number:
  print(Indvidual_Phone_num)
```

```python
print("Staff Details:")
for name, education, role, email,phone_no in zip(staff_names, staff_educations, staff_roles, staff_emails,staff_Phone_number):
    print("Name:", name)
    print("Education:", education)
    print("Desgination :", role)
    print("Email :", email)
    print("Phone number :",phone_no)
    print()
```

The above displayed the web scrapped data of faculty in a organized manner.

```python
import csv
#create CSV file
with open('staff_data.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Name', 'Education', 'Desgination', 'Email','Phone Number']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for name, education,role, email, phone_no in zip(staff_names, staff_educations, staff_roles,staff_emails, staff_Phone_number):
        writer.writerow({'Name': name, 'Education': education,'Desgination': role,'Email': email, 'Phone Number': phone_no})

print("CSV file created !!")
```

In above we have tries to create a Csv file to store scrapped data through importing the csv and opening the file discreptor in write mode inorder to write the data in named csv file "staff_data.csv".
